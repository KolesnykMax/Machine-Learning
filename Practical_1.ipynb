Завдання 1. Створіть одновимірний масив (вектор) з числами Фібоначчі, де кожне число дорівнює сумі двох попередніх, та виведіть його значення.

```python
n = 10
fibonacci = np.zeros(n, dtype=int)
fibonacci[0] = 0
fibonacci[1] = 1

for i in range(2, n):
    fibonacci[i] = fibonacci[i - 1] + fibonacci[i - 2]
print(fibonacci)
```

Завдання 2. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями експонент n^2 для кожного натурального числа n, починаючи з 1, та виведіть його значення.

```python
n = 3
matrix = np.zeros((n, n), dtype=float)

for i in range(n):
    for j in range(n):
      matrix[i][j] = np.exp((i * n + j + 1) ** 2)
print(matrix)
```

Завдання 8. Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю.

```python
matrix = np.random.randint(1, 16, (2, 2))

print("Початкова матриця:")
print(matrix)

try:
    inverse_matrix = np.linalg.inv(matrix)
    print("Обернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("Матриця не має оберненої матриці (детермінант дорівнює нулю).")
```

Завдання 15. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму квадратів елементів кожного рядка матриці.

```python
matrix = np.random.rand(3, 3)

print("Початкова матриця:")
print(matrix)

row_sums = np.sum(matrix**2, axis=1)

print("Сума квадратів елементів кожного рядка:")
print(row_sums)

column_vector = row_sums.reshape(-1, 1)

print("Вектор-стовпчик із сумами квадратів елементів кожного рядка:")
print(column_vector)
```
